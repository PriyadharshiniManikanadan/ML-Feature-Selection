## SelectK Algorithm - Classification

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
import pickle
```

```python
def selectkbest(indep_x,dep_y,n):
    SK = SelectKBest(score_func = chi2,k=n)
    SKfit = SK.fit(indep_x,dep_y)
    selectK_features = SKfit.transform(indep_x)
    selected_indices = SKfit.get_support(indices=True) 
    selected_features_names = indep_x.columns[selected_indices]
    return selectK_features, selected_features_names.tolist()

def split_scalar(indep_x,dep_y):
    x_train,x_test,y_train,y_test = train_test_split(indep_x, dep_y, test_size=0.2, random_state=0)
    sc = StandardScaler()
    x_train = sc.fit_transform(x_train)
    x_test = sc.transform(x_test)
    return x_train,x_test,y_train,y_test
```

```python
raw_dataset = pd.read_csv("CKD.csv",index_col=None)
raw_dataset
```

```python
df = raw_dataset
```

```python
indep_x = df.drop('classification_yes', axis = 1)
dep_y = df['classification_yes']
```

```python
kbest, selected_features_names = selectkbest(indep_x,dep_y,7)
```

```python
x_train,x_test,y_train,y_test = split_scalar(kbest,dep_y)
```

```python
# Define parameter grids for each classifier

params_grid = { 'logistic' : {'penalty' : ['l1', 'l2', 'elasticnet'],
                              'solver' : ['saga'],
                              'multi_class': ['auto', 'ovr', 'multinomial']},
               
              'svm_linear' : {'kernel':['linear'],
                               'C' : [1,10,100,1000,2000,3000]},
               
           'svm_nonlinear' : {'kernel':['linear','rbf','poly','sigmoid'],
                              'C' : [1,10,100,1000,2000,3000], 
                              'gamma' : ['scale','auto']},
               
                     'knn' : {'n_neighbors': [3, 5, 7],
                              'weights': ['uniform', 'distance']},

                    'naive': {},  # Naive Bayes has no hyperparameters to tune
               
                 'decision': {'criterion': ['gini', 'entropy'],
                              'max_depth': [None, 10, 20, 30]},

               'random': {'n_estimators': [50, 100, 200],
                              'max_depth': [None, 10, 20, 30],
                              'criterion': ['gini', 'entropy']}}               
               
```

```python
# Helper function to run GridSearchCV and return the best estimator and its accuracy

def run_gridsearch(estimator, params, x_train, y_train, x_test, y_test):
    grid_search = GridSearchCV(estimator, params, refit=True, verbose=3, n_jobs=-1, scoring='accuracy')
    grid_search.fit(x_train, y_train)
    best_model = grid_search.best_estimator_
    accuracy = best_model.score(x_test, y_test)
    return best_model, accuracy
```

```python
run_gridsearch(LogisticRegression(), params_grid['logistic'], x_train, y_train, x_test, y_test)
run_gridsearch(SVC(), params_grid['svm_linear'], x_train, y_train, x_test, y_test)
run_gridsearch(SVC(), params_grid['svm_nonlinear'], x_train, y_train, x_test, y_test)
run_gridsearch(KNeighborsClassifier(), params_grid['knn'], x_train, y_train, x_test, y_test)
run_gridsearch(DecisionTreeClassifier(), params_grid['decision'], x_train, y_train, x_test, y_test)
run_gridsearch(RandomForestClassifier(), params_grid['random'], x_train, y_train, x_test, y_test)
```
