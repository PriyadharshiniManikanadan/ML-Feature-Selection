```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import pickle
import matplotlib.pyplot as plt
```

```python
def split_scaler(indep_x,dep_y):
    x_train,x_test,y_train,y_test = train_test_split(indep_x,dep_y,test_size = 0.2,random_state = 0)
    sc = StandardScaler()
    x_train = sc.fit_transform(x_train)
    x_test = sc.transform(x_test)
    return x_train,x_test,y_train,y_test

def cm_prediction(classifier,x_test,y_test):
    y_pred = classifier.predict(x_test)

    from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
    cm = confusion_matrix(y_test,y_pred)
    accuracy = accuracy_score(y_test,y_pred)
    clas_report = classification_report(y_test,y_pred)
    return accuracy

def logistic(x_train,y_train,x_test):
    from sklearn.linear_model import LogisticRegression
    classifier = LogisticRegression()
    classifier.fit(x_train,y_train)
    accuracy = cm_prediction(classifier,x_test,y_test)
    return accuracy

def decision(x_train,y_train,x_test):
    from sklearn.tree import DecisionTreeClassifier
    classifier = DecisionTreeClassifier(random_state=0)
    classifier.fit(x_train,y_train)
    accuracy = cm_prediction(classifier,x_test,y_test)
    return accuracy    

def random(x_train,y_train,x_test):
    from sklearn.ensemble import RandomForestRClassifier
    classifier = RandomForestClassifier(n_estimators = 10, random_state=0)
    classifier.fit(x_train,y_train)
    accuracy = cm_prediction(classifier,x_test,y_test)
    return accuracy

def xgboost(x_train,y_train_,x_test):
    from xgboost import XGBClassifier
    classifier = XGBClassifier(n_jobs=5, learning_rate=0.01, max_depth=10, randon_state=1)
    classifier.fit(x_train,y_train)
    accuracy = cm_prediction(classifier,x_test,y_test)
    return accuracy
```

```python
def selectkbest(indep_x,dep_y,n):
    kbest_list = []
    cols_list = []
    Accuracy = []
    
    kbest = SelectKBest(score_func=chi2, k=n)
    kbest.fit(indep_x,dep_y)
    kbest_features = kbest.transform(indep_x)
    kbest_list.append(kbest_features)

    selected_features = [col for col, selected in zip(indep_x.columns, kbest.get_support()) if selected]
    cols_list.append(selected_features)

    from sklearn.linear_model import LogisticRegression
    log_model = LogisticRegression()

    from sklearn.tree import DecisionTreeClassifier
    dc_model = DecisionTreeClassifier(random_state=0)

    from sklearn.ensemble import RandomForestClassifier
    rf_model = RandomForestClassifier(n_estimators=10, random_state=0)

    from xgboost import XGBClassifier
    xgb_model = XGBClassifier(n_jobs=5, learning_rate=0.1, max_depth=10, random_state=1)

    modellist = [log_model, dc_model, rf_model, xgb_model]

    for model in modellist:
        x_train,x_test,y_train,y_test = split_scaler(pd.DataFrame(kbest_features), dep_y)
        model.fit(x_train,y_train)
        accuracies = cm_prediction(model,x_test,y_test)
        Accuracy.append(accuracies)
    
    return kbest_list, cols_list, Accuracy
```

```python
dataset = pd.read_csv("CKD.csv",index_col=None)
dataset
```

```python
df = dataset
```

```python
df = pd.get_dummies(df,dtype=int,drop_first=True)
```

```python
indep_x = df.drop('classification_yes',axis=1)
dep_y = df['classification_yes']
```

```python
kbest_list, cols_list, Accuracy = selectkbest(indep_x,dep_y,5)

for model_name, selected_features, accuracy in zip(["Linear", "Decision", "Random", "XGBoost"], cols_list, Accuracy):
    print(f"Model: {model_name}")
    print("Selected Features:", selected_features)
    print(f"Accuracy: {accuracy}\n")
```
